```python
from sklearn.datasets import load_digits
```

```python
X_digits, y_digits = load_digits(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_digits, y_digits)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression(multi_class="auto", solver="lbfgs", max_iter=10000)
log_reg.fit(X_train, y_train)
```

```python
log_reg.score(X_test, y_test)
```

```python
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ("kmeans", KMeans(n_clusters=50)),
    ("log_reg", LogisticRegression(multi_class="auto", solver="lbfgs", max_iter=20000))
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
pipeline.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = dict(kmeans__n_clusters=range(2,100))
grid_clf = GridSearchCV(pipeline, param_grid, cv=3, verbose=2)
grid_clf.fit(X_train, y_train)
```

```python
grid_clf.best_params_
```

```python
grid_clf.score(X_test, y_test)
```

```python
n_labeled = 50
log_reg = LogisticRegression()
log_reg.fit(X_train[:n_labeled], y_train[:n_labeled])
```

```python
log_reg.score(X_test, y_test)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(n_samples=1000, noise=0.05)
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=50)
knn.fit(dbscan.components_, dbscan.labels_[dbscan.core_sample_indices_])
```

```python
X_new = np.array([[-0.5, 0], [0, 0.5], [1, -0.1], [2, 1]])
```

```python
knn.predict(X_new)
```

```python
knn.predict_proba(X_new)
```

```python
y_dist, y_pred_idx = knn.kneighbors(X_new, n_neighbors=1)
```

```python
ypred = dbscan.labels_[dbscan.core_sample_indices_][y_pred_idx]
```
